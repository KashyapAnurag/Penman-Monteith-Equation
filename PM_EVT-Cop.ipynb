# PM-Evapotranspiration

```python
#||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>> PENMAN MONTEITH EQUATION CODE<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<#
#||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||
#--------------------------------------------------------------------------------------------------------------
# (1) PRESSURE
def calculate_pressure(elevation):
    return 101.3 * (((293 - 0.0065 * elevation) / 293) ** 5.26)
DATA['PRESSURE'] = DATA['ELEVATION'].apply(calculate_pressure)

#--------------------------------------------------------------------------------------------------------------

# (2) PSYCHROMETRIC CONSTANT (GAMMA)
def gamma_val(pressure):
    return 0.66 * 0.001 * pressure
DATA['GAMMA'] = DATA['PRESSURE'].apply(gamma_val)

#--------------------------------------------------------------------------------------------------------------

# (3) T_BAR
def TBAR(row):
    return (row['tmax'] - row['tmin']) / 2
DATA['T_BAR'] = DATA.apply(TBAR, axis=1)

#--------------------------------------------------------------------------------------------------------------

# (4) SATURATION VAPOUR PRESSURE
def SAT_VP_TBAR(T_BAR1):
    return 0.6108 * np.exp((17.27 * T_BAR1) / (T_BAR1 + 237))
DATA['e_TBAR'] = DATA['T_BAR'].apply(SAT_VP_TBAR)


#--------------------------------------------------------------------------------------------------------------

# (5) SLOPE: DELTA
def DELTA1(row):
    return ((4098 * row['e_TBAR'] ) / ((row['T_BAR'] + 237.3)**2))
DATA['DELTA'] = DATA.apply(DELTA1, axis=1)

###############################################################################################################
                                 #PART-A OF PENMAN MONTEITH EQUATION
    
DATA.reset_index(drop=True,inplace=True)
netRad2.reset_index(drop=True,inplace=True)

netRad2['PART_A'] = 0.408 * netRad2['netRad'] * DATA['DELTA']    #.mul(0.000001)

##############################################################################################################
                                 #PART-B OF PENMAN MONTEITH EQUATION

#--------------------------------------------------------------------------------------------------------------
# (6) e_tmax
def SAT_VP_TMAX(T_MAX):
    return 0.6108 * np.exp((17.27 * T_MAX) / (T_MAX + 237))
DATA['e_Tmax'] = DATA['tmax'].apply(SAT_VP_TMAX)

#--------------------------------------------------------------------------------------------------------------

# (7) e_tmin
def SAT_VP_TMIN(T_MIN):
    return 0.6108 * np.exp((17.27 * T_MIN) / (T_MIN + 237))
DATA['e_Tmin'] = DATA['tmin'].apply(SAT_VP_TMIN)

#--------------------------------------------------------------------------------------------------------------
# (8) actual vp
def actVAP(row):
    return (row['rhum'] * (row['e_Tmax'] + row['e_Tmin']) * 0.005)
DATA['e_act'] = DATA.apply(actVAP, axis=1)

#--------------------------------------------------------------------------------------------------------------
# (9) saturated vp
def satVAP(row):
    return ((row['e_Tmax'] + row['e_Tmin'])/ 2)
DATA['e_sat'] = DATA.apply(satVAP, axis=1)

#--------------------------------------------------------------------------------------------------------------
# (10) MEAN TEMPERATURE
def TMEAN(row):
    return (row['tmax'] + row['tmin']) / 2

# apply the T_BAR function to each row in the DataFrame and assign the results to a new column
DATA['T_MEAN'] = DATA.apply(TMEAN, axis=1)

#--------------------------------------------------------------------------------------------------------------
# (11) Merging the wind data 

# wind_2m.reset_index(drop=True,inplace=True)       #wind data
wind_2 = wind_2m[['wind_2m']]

DATA = pd.merge(DATA,wind_2, left_index=True, right_index=True)

#--------------------------------------------------------------------------------------------------------------
# (12) m_part1
def partM(row):
    return (row['GAMMA'] * (900/(row['T_MEAN'] + 273)) * (row['e_sat'] - row['e_act']))
DATA['B_1'] = DATA.apply(partM, axis=1)

#--------------------------------------------------------------------------------------------------------------
# (13) Part-B
netRad2['PART_B'] =  DATA['wind_2m'] * DATA['B_1']

##############################################################################################################
                                 #PART-C OF PENMAN MONTEITH EQUATION

#--------------------------------------------------------------------------------------------------------------
netRad2['PART_C'] =  DATA['DELTA'] + ( DATA['GAMMA'] * (1 + 0.34 * DATA['wind_2m']))


##############################################################################################################
                         #REFERENCE EVAPOTRANIRATION PENMAN MONTEITH EQUATION

netRad2['ET_node'] = (netRad2['PART_A'] + netRad2['PART_B']) / netRad2['PART_C'] 

##############################################################################################################
##############################################################################################################
##############################################################################################################
```
